```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#Dataset for electricity consumption forecasting
data = {
    "Temperature": [32, 46, 25, 29, 52, 40, 38],
    "Humidity": [55, 60, 58, 62, 65, 59, 61],
    "Day_of_Week": [0, 1, 2, 3, 4, 5, 6],
    "Electricity_Consumption": [123, 235, 346, 457, 568, 490, 510]
}

#Accessing dataframes through pandas
df = pd.DataFrame(data)
print(df)


#Visualizations

# Bar Chart
plt.bar(df["Day_of_Week"], df["Electricity_Consumption"])
plt.title("Electricity Consumption by Day")
plt.xlabel("Day of Week")
plt.ylabel("Consumption")
plt.show()

X = df[["Temperature", "Humidity", "Day_of_Week"]]
y = df["Electricity_Consumption"]

#Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

#Train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

#Prediction
y_pred = model.predict(X_test)

#Evaluation
print("Actual values:",list(y_test.values))
print("Predicted values:",list(y_pred))

# Actual vs Predicted Plot
plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label="Actual", marker="o")
plt.plot(y_pred, label="Predicted", marker="x")
plt.title("Actual vs Predicted Electricity Consumption")
plt.legend()
plt.show()
```
